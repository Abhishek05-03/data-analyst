```python
# creating a Series from a list
import pandas as pd
data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print(series)
```

```python
# create a DataFrame from a dictionary
data = {"a" :1 , "b": 2, "c": 3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data2 = [1,2,3,4,5]
index = ['a', 'b', 'c', 'd', 'e']
series_index = pd.Series(data2, index=index)
print(series_index)
```

```python
# Data Frame creation
# create a DataFrame from a dictionary of lists
data = {
    "Name": ["Peter Pakar", "Tony Stark", "Bruce Wayne"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Create A DataFrame from a list of dictionaries
data_list = [
    {"Name": "Peter Pakar", "Age": 25, "City": "New York"},
    {"Name": "Tony Stark", "Age": 30, "City": "Los Angeles"},
    {"Name": "Bruce W ayne", "Age": 35, "City": "Chicago"}
]
df_list = pd.DataFrame(data_list)
print(df_list)  
```
