```python
data = pd.read_csv("data.csv") # Reading a CSV file into a DataFrame
df = pd.DataFrame(data) # Creating a DataFrame from the read CSV data
df.tail(5) # Displaying the last few rows of the DataFrame
```

```python
df = df.rename(columns = {"Date" : "Sales date"})   # Renaming the "date" column to "Sales date"
df.head()
```

```python
# Change the data type 

df['value_new'] = df['Value'].fillna(df['Value'].mean()).astype(int) # Changing the data type of the "Value" column to float

df.head() # Displaying the first 5 rows of the DataFrame
```

```python
df['new_values'] = df["Value"].apply(lambda x: x * 2) # Applying a lambda function to double the values in the "Value" column and creating a new column "new_values"
df.head() # Displaying the first 5 rows of the DataFrame
```

```python
# data Aggregating And Grouping 
df.head()
```

```python
grouped_mean = df.groupby("Product")["Value"].mean() # Grouping the DataFrame by "Category" and calculating the mean of the "Value" column for each group
print(grouped_mean)
```

```python
grouped_sum = df.groupby(["Product", "Region"])["Value"].sum() # Grouping the DataFrame by "Category" and calculating the sum of the "Value" column for each group
print(grouped_sum)
```

```python
grouped_mean = df.groupby(["Product", "Region"])["Value"].mean() # Grouping the DataFrame by "Category" and calculating the sum of the "Value" column for each group
grouped_mean
```

```python
# Aggregating multiple functions
grouped_agg = df.groupby("Product")["Value"].agg(["mean", "sum", "count"]) # Grouping the DataFrame by "Category" and applying multiple aggregation functions (mean, sum, count) to the "Value" column
grouped_agg
```

```python
# Merging And joining  DataFrames
# create sample DataFrames
df1 = pd.DataFrame({'key' : ['A', 'B', 'C'], 'value1': [1, 2, 3]})
df2 = pd.DataFrame({'key' : ['A', 'B', 'D'], 'value2': [4, 5, 6]})    
```

```python
df1
```

```python
df2
```

```python
# Merging DataFrames on the key column
merged_df = pd.merge(df1, df2, on='key', how='inner')
merged_df
```

```python
merged_df = pd.merge(df1, df2, on='key', how='outer')
merged_df
```

```python
merged_df = pd.merge(df1, df2, on='key', how='left')
merged_df
```

```python
merged_df = pd.merge(df1, df2, on='key', how='right')
merged_df
```
