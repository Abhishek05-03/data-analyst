```python
# Set MemberShip test 
my_set = {1,2,3,4,5}
print(3 in my_set)
print(6 in my_set)
```

```python
#math operations on set
A = {1,2,3,4,5}
B = {4,5,6,7,8}

# union of sets
union_set = A.union(B)
print(union_set) # union of sets is the set of all elements that are in either A or B or in both A and B.

# intersection of sets
intersection_set = A.intersection(B)
print(intersection_set) # intersection of sets is the set of elements that are common to both A and B.

# Difference of sets
set_difference = A.difference(B)
print(set_difference) # difference of sets is the set of elements that are in A but not in B.


set_difference = B.difference(A)
print(set_difference)

#symmetric difference of sets
symmetric_difference = A.symmetric_difference(B)
print(symmetric_difference) # symmetric difference of sets is the set of elements that are in either A or B but not in both.

```

```python
# set Methods

set1 = {1,2,3,4,5}
set2 = {3,4,5}

print(set1.issubset(set2))
print(set1.issuperset(set2))
```
