##### SMOTE(synthetic Minority Oversampling technique)
instead of simply copying minority samples (like basic oversampling), SMOTE:

```python
from sklearn.datasets import make_classification
x,y =make_classification(n_samples =1000 , n_redundant=0,n_features=2,n_clusters_per_class=1,
weights=[0.90],random_state=12)
```

```python
import pandas as pd
df1 = pd.DataFrame(x,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns =['target'])
final_df = pd.concat([df1,df2],axis =1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

### 5 Number Summary And Box Plot

```python
import numpy as np
```

```python
lst_number = [45,32,56,54,75,32,13,47,45,33,26,31,34]
mimimum , Q1,median,Q3,maximum = np.quantile(lst_number,[0,0.25,0.50,0.75,1.0])
print(mimimum,Q1,median,Q3,maximum)
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
lower_fence = Q1-1.5*(IQR)
Higher_fence = Q3+1.5*(IQR)
```

```python
print("Lower Fence" ,lower_fence)
print(f"Higher Fence   {Higher_fence}")
```

```python

import seaborn as sns
sns.boxplot(x=lst_number)
```
