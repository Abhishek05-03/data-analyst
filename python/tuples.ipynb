```python
# packing and unpacking of tuples

packed_tuple = 1,2,3,4,5,6,7,8,9
print(packed_tuple)
```

```python
# unpacking of tuples
a,b,c,d,e ,f,g,h,i= packed_tuple
print(a)
print(b)
print(c)
print(d)
print(e)
print(i)
```

```python
# unpacking with * operator
a,b,c,*d = packed_tuple
print(a)
print(b)
print(c)
print(d)
```

```python
a,b,*c,d = packed_tuple
print(a)
print(b)
print(c)
print(d)
```

```python
# nested tuples
nested_tuple = (1,2,3,(4,5,6),7,8,9)
print(nested_tuple)
print(nested_tuple[3])
print(nested_tuple[3][0])
```

```python
natural_numbers = tuple(i for i in range(1,11))
print(natural_numbers)
```

```python
print(f"First element of natural_numbers: {natural_numbers[0]}")
print(f"Last element of natural_numbers: {natural_numbers[-1]}")
print(f"middle Element of natural_numbers: {natural_numbers[len(natural_numbers)//2]}")
```
