```python
student = {
    "name": "Abhishek",
    "age" : 20,
     "branch" : "CSE(AI)"

}
print(student)
```

```python
# accessing dictonary elements
print(student["name"])
```

```python
print(student["age"])
```

```python
# Accessing using get method
print(student.get("name"))
print(student.get("age"))
print(student.get("roll_no")) # if key is not present it will return None
print(student.get("roll_no","Key not found")) # if key is not present it will return the default value
```

```python
# modyfing dictonary values
student["age"] = 21 # update the value of age key to 21
print(student)
student["roll_no"] = 101 # add a new key-value pair to the dictonary
print(student)
```

```python
del student["roll_no"] # delete the key-value pair with key "roll_no"
print(student)
```

```python
#dictonary methods

keys = student.keys() # get all keys
print(keys)

values = student.values() # get all values
print(values)

items = student.items() # get all key value pair
print(items)
```

```python
# Shallow copy
student_copy =  student.copy()
print(student_copy)
```

```python
student["name"] ="aman"
print(student)
print(student_copy)
```

```python
student_copy =  student
print(student_copy)
print(student)

student["name"] = "Abhishek"
print(student_copy)
print(student)

# in this method change value also in copy dict
```

```python
# itration using for loops

#iterate over keys

for keys in student.keys():
    print(keys)
```

```python
for values in student.values():
    print(values)
```

```python
for keys ,values in student.items():
    print(f"{keys} : {values}")
```

```python
# Nested dictonary 

student = {
    "student1" : {"name" : "spiderman" , "age" : 23},
    "student2" : {"name" : "superman" , "age" : 35}

}
print(student)
print(student["student2"])
print(student["student2"]["name"])
```

```python
# itrateing over nested dictonaries
for student_id , student_info in student.items():
    print(f"{student_id} : {student_info}")
    for keys , values in student_info.items():
        print(f"{keys} : {values}")
```

```python
# dictonary comprehension
squares = {x : x**2 for x in range(1,11)}
print(squares)
```

```python
# conditional dictonary comprehension
even_squares = {x : x**2 for x in range(1,11) if x % 2 == 0}
print(even_squares)
```
