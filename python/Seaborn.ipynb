```python
import seaborn as sns
```

```python
# Basic Plotting With Seaborn
tips = sns.load_dataset('tips')
tips
```

```python
# Create a scatter Plot

import matplotlib.pyplot as plt
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot of total bill VS tip")
plt.show()

```

```python
# Line Plot 
import matplotlib.pyplot as plt
sns.lineplot(x='size', y='total_bill' , data =tips)
plt.title("Line plot of Total bills VS Size")
plt.show()
```

```python
sns.barplot(x='size', y='total_bill' , data=tips)
plt.title(' Bar Plot Total Bill VS Day')
plt.show()
```

```python
# box plot
sns.boxplot(x='size',y='total_bill', data=tips)
```

```python
# vioin plot

sns.violinplot(x='size' , y="total_bill" , data=tips)
```

```python
# histograms 

sns.histplot(tips['total_bill'] ,bins=10,kde=True)
```

```python
#KDE Plot

sns.kdeplot(tips['total_bill'], fill=True)
```

```python
# pair Plot
sns.pairplot(tips)
```

```python
# Heat map 
corr =tips[['total_bill','tip','size' ]]
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd 
info_df = pd.read_csv('tips.csv')
info_df
```

```python
# Plot total bill by custember

plt.figure(figsize=(10,6))
sns.barplot(x='total_bill' , y='tip',data=info_df,estimator=sum)
plt.title('Plot total bill by custember')
plt.xlabel('Total Bills')
plt.ylabel('Tips')
plt.show()

```
