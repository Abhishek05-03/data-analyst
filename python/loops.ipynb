```python
# While loops 

# while loops is used to execute a block of code repeatedly until a certain condition is met. The syntax for a while loop is as follows:

count = 0

while count <=5:
    print(count)
    count +=1
```

```python
# loops condition statement

#Break - break statement exist the loops prematurely

#break statement 

for i in range(10):
    if i== 6:
        break
    print(i)
```

```python
# continue  - the continue statement skips the current iteration with the next.


for i in range(10):
    if i== 6:
        continue
    print(i)
```

```python
# pass - pass statement is a null operation it does nothing
for i in range(10):
    if i%2==0:
        pass
    print(i)

```

```python
## nested loops 
 
 
for i in range(3):
    for j in range(2):
        print(f" i is {j} and j is {j}")
```

```python
# calc sum of natural number

# while loop
n = int(input("Enetr A number : "))
sum = 0
count =0

while count <= n:
     sum= sum+count
     count+=1
print(f"sum of natural No {n} is : {sum}")   
```

```python
# for loop 
n = int(input("Enetr A number : "))
sum = 0

for i in range(n+1):
    sum = sum +i
 
print(f"sum of natural No is : {sum}")   
```

```python
# prime No 
n = int(input("Enetr A number : "))
for num in range(2,n+1):
    for i in range(2,num):
        if num%i == 0:
            break
    else:
        print(num ,end=" ")
    
```
